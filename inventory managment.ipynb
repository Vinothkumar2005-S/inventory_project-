```python
# change the datatypes 
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
df.dtypes
```

```python
## data cleaning 
df.Category.isna().sum()
df.isna().sum()
df.isnull().sum()
```

```python
# remove duplicates
df.duplicated().sum()
```

```python
df.drop_duplicates()
```

```python
df.columns = df.columns.str.lower()
df.columns
```

```python
df1 = df.rename(columns={'store id':'store_id','product id':'product_id','inventory level':'inventory_level','units sold':'units_sold','units ordered':'units_ordered','demand forecast':'demand_forecast','weather condition':'weather_condition','competitor pricing':'competitor_pricing'})
df1
```

```python
# adding columns to table 
df1['profits'] = df1['units_sold'] * df1['price']
df1['profits']
```

```python
## apply functions 

df1['price_double'] = df1['competitor_pricing'] + df1['price']
df1['price_double']
```

```python
## using some functions 

def forcast(demand_forecast):
    return demand_forecast * 2

df1['squared_forecast'] = df1['demand_forecast'].apply(forcast)
df1

```

```python
## filtering functions

df1.nlargest(n = 5, columns='profits')
```
